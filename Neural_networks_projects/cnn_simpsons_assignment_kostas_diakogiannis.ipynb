```python
from keras import layers, models, optimizers
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from keras.applications import VGG16
```

## Very Important Notes:

This dataset originally contains more images, though the structure had to be removed, and built manually from the beginning.

The reason is because the images that were inside the test folder of each class, were already in the train folder, that would cause an false-sense of success and measured not as fair. Thus the images in test folder were emptied for each class and then randomly filled with 20% of the train set images.

Same happened with validation set. The way I constructed the whole image folder structure it was with the help of a script, that I wrote in another jupyter notebook locally. Hopefully, I was able to upload the image structure as zipped and it remained by that.

```python
# normalize our set

train_datagen = ImageDataGenerator(rescale=1./255)
validation_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)
```

### Classification Details

The original dataset (which by the way you can find it here)[https://www.kaggle.com/alexattia/the-simpsons-characters-dataset] if you want to download and have a look at the original structure, consists of 42 classes-characters.

Becuase the distribution though is highly imbalanced and some characters contain a bigger portion of the set (like bart and homer) we filtered them by popularity and focused to the ones whose training and test image set is a significant number. 

1. So the classes we will examine will be 18
2. The images will be served as an input of (128, 128) because the images have not all the same size and I wouldn't like to have unlucky crops or other unwanted accidents.

That being said, let's proceed!

```python
train_generator = train_datagen.flow_from_directory(
    'simpsons_set/train/',
    target_size=(128, 128),
    batch_size=64,
    class_mode='categorical'
)

validation_generator = validation_datagen.flow_from_directory(
    'simpsons_set/validation/',
    target_size=(128, 128),
    batch_size=64,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    'simpsons_set/test/',
    target_size=(128, 128),
    batch_size=64,
    class_mode='categorical',
    shuffle=False
)
```

```python
cnn_model = models.Sequential()

# we take an input image of 3 dimensions and 128 X 128 width and height, at the beginning we have 32 filters, and 3x3 window
cnn_model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
cnn_model.add(layers.MaxPooling2D((2,2)))

cnn_model.add(layers.Conv2D(64, (3, 3), activation='relu'))
cnn_model.add(layers.MaxPooling2D((2, 2)))

cnn_model.add(layers.Conv2D(128, (3,3), activation='relu'))
cnn_model.add(layers.MaxPooling2D((2,2)))

cnn_model.add(layers.Conv2D(128, (3,3), activation='relu'))
cnn_model.add(layers.MaxPooling2D((2,2)))

cnn_model.add(layers.Flatten())
cnn_model.add(layers.Dense(512, activation='relu'))
cnn_model.add(layers.Dense(512, activation='relu'))
cnn_model.add(layers.Dense(18, activation='softmax'))
```

```python
cnn_model.summary()
```

```python
cnn_model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
train_images = 10974
validation_images = 3682

train_steps_per_epoch = train_images // 64
validation_steps = validation_images // 64
```

```python
history = cnn_model.fit_generator(
    train_generator, 
    steps_per_epoch=train_steps_per_epoch, 
    epochs=20, 
    validation_data=validation_generator, 
    validation_steps=validation_steps
    )
```

```python
cnn_model.save('final_cnn.h5')
```

```python
cnn_loss, cnn_score = cnn_model.evaluate(test_generator)

cnn_score
```

```python
# plot the differences

epochs = np.arange(1, len(history.history['val_loss']) + 1)

plt.figure(figsize=(10, 5))
plt.title('Accuracy difference per epoch')
plt.plot(epochs, history.history['accuracy'])
plt.plot(epochs, history.history['val_accuracy'])
plt.xlabel('epoch iteration')
plt.xticks(ticks=epochs, labels=epochs)
plt.ylabel('Model accuracy')
plt.legend('upper left', labels=['train_acc', 'val_acc'])
plt.show()


plt.figure(figsize=(10, 5))
plt.title('Loss difference per epoch')
plt.plot(epochs, history.history['loss'])
plt.plot(epochs, history.history['val_loss'])
plt.xlabel('epoch iteration')
plt.ylabel('Model loss')
plt.xticks(ticks=epochs, labels=epochs)
plt.legend('upper right', labels=['train_loss', 'val_loss'])
plt.show()
```

```python
# predict the labels

predictions = cnn_model.predict_generator(test_generator)

# take the most probable class
predicted_labels = np.argmax(predictions, axis=1)
```

#### Very quick note:

So we can analyse (in a while a bit more in detail), and we can see that for **grandpa abraham simpson** the most common classification is **homer simpson**. Maybe it's in the mouth, which seems to be the same, who knows? 

```python
label_indices = train_generator.class_indices
actual_labels = dict((v, k) for (k, v) in label_indices.items())
actual_labels
```

```python
predictions_mapped = [actual_labels[x] for x in predicted_labels]
predictions_mapped[0:30]

# now it's even more clear than homer is the threat for abraham grandpa simpson
```

```python
# convert digit classes to string character's names
labels_mapped = [actual_labels[x] for x in test_generator.classes]
```

```python
# print classification report 
report = classification_report(labels_mapped, predictions_mapped, target_names=label_indices.keys())
print(report)
```

### Classification report notes

I would definitely expect `Homer` to be widely recognise and have a higher score, since we have also much more photos of him, due to his popularity. Someone would expect that the characters that appear more frequently, should have `get acquiainted` with the model. 

That doesn't seem to be the way though, since homer and grandpa mostly confuse our model. 

On the other hand more `unique` physiques like the ones of **milhouse van houten** and **marge simpson** seem to be easily recognized.

At last, **Lisa** and **Bart** which are also main characters of the show and we have gothered lots of samples, still fail miserably. Let's see with which characters does the model confuse them.

```python
cf_matrix = confusion_matrix(labels_mapped, predictions_mapped, normalize='true')
plt.figure(figsize=(15, 7))
sns.heatmap(cf_matrix, annot=True, cmap='vlag', xticklabels=actual_labels.values(), yticklabels=actual_labels.values())
plt.title('Confusion Matrix for simpsons cnn model')
plt.show()
```

### Confusion matrix heatmap notes

Now we can see not only who was predicted more accurately, but also with whom was he/she mistaken for.

For example is easy to detect the characters that were more or less acurrately predicted like:

1. Krusty the clown
2. Abraham grandpa
3. Sideshow Bob 
4. Kent Brockman

But also the obvious cases of underperforming like:

1. Lenny leonard, who was more oftenly mistaken either by `Homer` or by `Abraham grandpa`
2. Nelson Muntz, who had a series of character missclassifications with bart being the strongest.

An interesting fact `Bart` and `Lisa` Simpsons, are the most frequent cases of missclassifications of one another and they are siblings. Good Job Simpsons, they do really look like :-). Actually Bart more than Lisa to be honest. **20%** of the time Lisa's character was passed for Bart. While in case of Bart, the percentage falls to **9%**.

```python
# Let's use a pre trained model to see what happens

pre_train_base = VGG16(weights='imagenet', include_top=False, input_shape=(128, 128, 3))
pre_train_base.summary()
```

```python
# we will add our own dense MLP at the end of it now

cnn_model = models.Sequential()
cnn_model.add(pre_train_base)
cnn_model.add(layers.Flatten())
cnn_model.add(layers.Dropout(0.2))
cnn_model.add(layers.Dense(512, activation='relu'))
cnn_model.add(layers.Dense(512, activation='relu'))
cnn_model.add(layers.Dense(18, activation='softmax'))
```

```python
cnn_model.summary()
```

```python
# freeze weights from getting updated

pre_train_base.trainable = False
```

```python
cnn_model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```
