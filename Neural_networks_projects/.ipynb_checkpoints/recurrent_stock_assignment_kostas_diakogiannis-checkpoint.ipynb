```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, LSTM, SimpleRNN, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
stock_set = pd.read_csv('apple_stock_data.csv')
stock_set.head()
```

```python
stock_values = stock_set['Close/Last'].apply(lambda x: x[1:]).astype('float')
stock_values[0:10]

reversed_stocks = stock_values[::-1]
```

```python
plt.figure(figsize=(15, 8))
plt.plot(reversed_stocks.index[::-1], np.array(reversed_stocks))
plt.show()
```

```python
# prepare data for recurrent neural network
#take reverse order since we want the most relevant to be the recent ones, and they will get fed in the end.

stocks = stock_values.values[::-1] 
stocks[0:5]
```

```python
stocks_reshaped = stocks.reshape(-1, 1)
stocks_reshaped
```

```python
# scale input data, we are still talking about ANN's!

scaler = MinMaxScaler()
stocks_scaled = scaler.fit_transform(stocks_reshaped)
stocks_scaled
```

```python
# plot the scaled version

plt.figure(figsize=(15, 8))
plt.plot(reversed_stocks.index[::-1], np.array(stocks_scaled))
plt.show()
```

```python
# we will set the steps every 30 days, like month by month basis, so at first iteration days 1-30 will be the X_train and day 31
# will be the y_train, during  next value, the sequence between days 2-31 will be X_train and day 32 the y_train and so on,
# we could predict also something more meaningful like the average value of the next 5 days, but for now keep it simple.

day_steps = 30 # evaluating lost month, predicting next day
total_days = stock_set.shape[0]

train_days = round(total_days * 0.8)
test_days = total_days - train_days

X_total = [stocks_scaled[i - day_steps: i, 0] for i in range(day_steps, total_days)]
y_total = stocks[day_steps:total_days]

X_train = X_total[0:train_days]
y_train = y_total[0: train_days]

X_test = X_total[train_days:]
y_test = y_total[train_days:]

# convert to numpy array and do some proper reshape before input

X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1)) # input preprocessing
X_train.shape
```

```python
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1] ,1))
X_test
```

```python
rnn_model = Sequential()

rnn_model.add(LSTM(16, input_shape=(30, 1)))
rnn_model.add(Dense(1))
rnn_model.compile(optimizer='adam', loss='mean_squared_error')
rnn_model.fit(X_train, y_train, epochs=15 , batch_size=1)
```
