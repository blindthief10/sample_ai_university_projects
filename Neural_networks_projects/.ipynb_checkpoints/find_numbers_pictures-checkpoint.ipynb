```python
import pandas as pd
```

```python
initial_set = pd.read_csv('number_pic_char.csv')
initial_set
```

```python
# set train 60% train, 20% validation set, 20% test

initial_set['train'] = initial_set['total'] * 0.6
initial_set['test'] = initial_set['total'] * 0.2
initial_set['validation'] = initial_set['test']

initial_set['train'] = initial_set['train'].astype('int32')
initial_set['test'] = initial_set['test'].astype('int32')
initial_set['validation'] = initial_set['validation'].astype('int32')

initial_set.drop(columns='bounding_box', inplace=True)
```

```python
initial_set.drop(columns='Unnamed: 0', inplace=True)
```
